```python
import numpy as np
import wfdb as wf
import pandas as pd
```

```python

ECG_folder_path = '/home/hsiehch/dataset/'
filename = 'Record_5309'

index = wf.rdsamp(ECG_folder_path + filename)
record = index[0]
record.shape = (record.shape[0], )
print(record)
```

```python
import pylab as plt

plt.figure(figsize=(30,4))
plt.plot(record)
plt.show()
```

```python
from biosppy.signals import ecg
peaks = ecg.christov_segmenter(record, 300)
print(peaks[0])
peaks2 = ecg.engzee_segmenter(record, 300)
print(peaks2[0])
```

```python
qrs = np.full(record.shape[0], None)
for i in peaks[0]:
    qrs[i] = record[i]
print(qrs)
```

```python
plt.figure(figsize=(30,4))
plt.plot(record)
plt.plot(qrs, 'ro')
plt.show()
```

```python
qrs2 = np.full(record.shape[0], None)
for i in peaks2[0]:
    qrs2[i] = record[i]
print(qrs2)
```

```python
plt.figure(figsize=(30,4))
plt.plot(record)
plt.plot(qrs2, 'ro')
plt.show()
```
