```python
import numpy as np
import pprint
from scipy.signal import spectrogram
import pylab as plt
np.set_printoptions(threshold=np.nan)

def to_Spectrogram(file_name, data, ratio=2, save_png=False, draw_png=True, print_data=False:
    f, t, Sx = spectrogram(data, fs=300, nperseg=64, noverlap=32, detrend=False, scaling="spectrum", mode="magnitude")
    Sx = np.log10(Sx)
    print(Sx.shape)
    if print_data:
        pprint.pprint(Sx)
    if draw_png:
        plt.figure(figsize=(Sx.shape[1], Sx.shape[0]))
        plt.pcolormesh(t, f, Sx)
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.show()
    if save_png:
        plt.imsave('/home/hsiehch/30s/%s.png' % file_name, Sx, origin="lower")

trainD = np.load("/home/hsiehch/30s/train_data.npy")
to_Spectrogram('file', trainD[0])
# validationD = np.load("/home/hsiehch/30s/validation_data.npy")
# testD = np.load("/home/hsiehch/30s/test_data.npy")

# trainD = np.append(trainD, validationD, axis=0)
# trainD = np.append(trainD, testD, axis=0)

# for i in range(len(trainD)):
#     to_Spectrogram(i, trainD[i])
```
