```python
import numpy as np
import pylab as plt
```

```python
data = np.load('/home/hsiehch/30s/train_data.npy')
data.shape
```

```python
trainDataIndex = []
for i in range(len(data)):
    for j in range(1, len(data[i])):
        if abs(data[i][j-1] - data[i][j]) > 1:
            trainDataIndex.append(i)
            break
print(trainDataIndex)
```

```python
plt.figure(figsize=(30,5))
plt.xlim(0,9000)
plt.plot(data[13])
plt.show()
```

```python
from sklearn.preprocessing import normalize
```

```python
tmpData = normalize(data[13].reshape(1,9000), norm='l2')
plt.figure(figsize=(30,5))
plt.xlim(0,9000)
plt.plot(tmpData[0])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(-4,2))
scaler.fit(data[13].reshape(1,9000))
tmpData = scaler.transform(data[13].reshape(1,9000))
plt.figure(figsize=(30,5))
plt.xlim(0,9000)
plt.plot(tmpData[0])
plt.show()
```
